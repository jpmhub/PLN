```python
# Lectura de fitxers de texte pla en local
import os
# podem posar la ruta absoluta o relativa
path = 'C:\\Users\\anton\\OneDrive\\Escriptori\\MODELS-IA\\TEMA3\\ejemplos-txt'
ficheros = os.listdir(path)

documentos = []

for nombrefichero in ficheros:    
        if os.path.isfile(os.path.join(path, nombrefichero)): #si és un fitxer i no és un directori
            fich = open(os.path.join(path, nombrefichero), "r")
            text = fich.read()
            documentos.append(text)
            
print(documentos)
```

```python
# Lectura de webs
get_ipython().system('pip install beautifulsoup4')
import urllib.request

from bs4 import BeautifulSoup

paginaweb= 'https://paucasesnovescifp.cat/'
with urllib.request.urlopen(paginaweb) as response:
   html = response.read().decode() # alerta si està ben codificada o no la web

soup = BeautifulSoup(html, features="html.parser")

# si volem netejar un poc
#import re
#print(re.sub('<[^<]+?>', '', html))

# Borra elements de script i style 
for script in soup(["script", "style"]):
    script.extract()    # borrar    
print(soup.get_text())
```
